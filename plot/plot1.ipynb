```python
import pandas as pd
data = pd.read_csv("ny_state_school3.csv")
data.columns
```

```python
data['PI'] = data['Principal_Investigator_1_First_Name'] + ' ' + data['Principal_Investigator_1_Last_Name']
```

```python
data1 = data.groupby(['PI'])['Total_Amount_of_Payment_USDollars'].sum().reset_index()
data2 = data.drop(['Total_Amount_of_Payment_USDollars'], axis = 1).merge(data1, how = 'left', on = 'PI')
```

```python
data = data2.sort_values(by = ['Total_Amount_of_Payment_USDollars'], ascending = False)[['School','Submitting_Applicable_Manufacturer_or_Applicable_GPO_Name','PI','Total_Amount_of_Payment_USDollars']]
```

```python
data = data.drop_duplicates()
data = data.dropna()
data['PI'] = data['PI'].map(lambda x: " ".join([i[0].upper()+i[1:].lower() for i in x.split(" ") if type(i) is str]))
data.columns = ['School','Company','PI','Money']
data.head()
data1 = data
data = data1.iloc[:200]
```

```python
names = list(data['PI'].unique())
outputs = []
for name in names:
    output1 = list(data[data['PI'] == name]['Company'].unique())
    output2 = list(data[data['PI'] == name]['School'].unique())
    outputs = outputs + [[int(data[data['PI'] == name]['Money'].iloc[0]/1000),output1,output2,name]]
outputs
```
